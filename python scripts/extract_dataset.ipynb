```python
import pandas as pd
```

```python
df = pd.read_csv("../datasets/option_GS_df.csv", index_col=0)
```

```python
# rename  DataDate to DataDate
df.rename(columns={' DataDate': 'DataDate'}, inplace=True)
```

```python
# Preprocessings to match the implementation in C++
df['Type'] = df['Type'].apply(lambda x: 1 if x == 'call' else 0)
df['Expiration'] = pd.to_datetime(df['Expiration'])
df['DataDate'] = pd.to_datetime(df['DataDate'])
```

```python
# calculate the time to expiration
df['TTE'] = (df['Expiration'] - df['DataDate']).dt.days / 365
```

```python
# save the Strike column to a txt file
df["Strike"].to_csv("strike.txt", header=False, index=False)
df['UnderlyingPrice'].to_csv("option_price.txt", header=False, index=False)
df['Type'].to_csv("type.txt", header=False, index=False)
df['TTE'].to_csv("tte.txt", header=False, index=False)
```
